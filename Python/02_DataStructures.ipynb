```python
def process_b(lines):
    result=[]
    
    for line in lines:
        if len(line)>1:
            
            result.append(line[0])
    return result

process_b(lines)
    
```

```python
def process(f,lines):
    result=[]
    
    for line in lines:
        if len(line)>1:
            
            result.append(f(line))
    return result

process(len,lines)
```

```python
names=['adria','florki','edu','berts','jose']

new_names=[]
def put_in_capital (nombres):
    
    for nombre in nombres:
        answer=""
        for letter in nombre:
            if letter  in "aeiou":
                answer = answer + letter.upper()
            else:
                answer = answer + letter.lower()
           
        new_names.append(answer)
    return new_names
put_in_capital(names)
        
    
```

```python
apellidos=['cespedes','almagro','sanchez','andara']
new_apellidos=[]
def filtro(nombres,character):
    
    for nombre in nombres:
        if character in nombre:
            new_apellidos.append(nombre)
        else:
            continue
    return(new_apellidos)
   
    

    

filtro(apellidos,'a')
        
        
```

```python
def orden_inverso(nombres):
    new_names=[]
    for nombre in nombres:
        nombre_prov=list(nombre)
        answer=""
        z=len(nombre_prov)
        while z>0:
            answer=answer + nombre_prov[(z-1)]
            z=z-1
        new_names.append(answer)
    return new_names

orden_inverso(apellidos)
            
        
            
        
```

```python
def maxima_longitud(nombres):
    longitud=[0]
    
    for nombre in nombres:
        nombre_prov=list(nombre)
        x=len(nombre_prov)
        if x>longitud[0]:
            longitud[0]=x
        else:
            continue
    return longitud[0]

maxima_longitud(apellidos)
        
    
```
